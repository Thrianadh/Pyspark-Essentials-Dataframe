Reading the dataset

```
### Option 1
SC_df = spark.read.option("header","true").csv("/FileStore/tables/practice1.csv",inferSchema=True )
print("="*50)
print(SC_df.show())
print("="*50)
### Option 2
SC_df = spark.read.csv("/FileStore/tables/practice1.csv",header= True,inferSchema=True)
print("="*50)
print(SC_df.show())
print("="*50)
#SC_df.printSchema()
print(type(SC_df.printSchema()))

```

Checking the datatypes of the column and describe

```
print(SC_df.columns)                      # getting the columns names

SC_df.head()                              # the top records
SC_df.head(2) 
SC_df.head(3)

SC_df.tail(1) # Diplay the bottom records
SC_df.tail(3) 


type(SC_df["Name"])

SC_df.dtypes                              # Checking the datatypes
SC_df.printSchema()                       # Checking the print schema
SC_df.describe().show()                   # Siply the the summary report
```

selecting columns and indexing

```
SC_df.select("Name").show()               # Selecting Single Column
SC_df.select(["Name","Age"]).show()       # Selecting Multiple Columns
```

Adding Columns

```
SC_df = SC_df.withColumn("Age after two Years", SC_df["Age"]*2) # Adding the column
SC_df.show()
```

Droping the column

```
SC_df = SC_df.drop("Age after two Years") # dropping the column
SC_df.show()
```

Column renamed

```
SC_df = SC_df.withColumnRenamed("Age","AGE") # columns renamed
SC_df.show()
```
